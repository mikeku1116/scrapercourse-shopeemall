```python
import pandas as pd
from snownlp import SnowNLP

data = pd.read_csv('shopeemall.csv')
data
```

```python
semantic = []
for row in data.itertuples():
    s = SnowNLP(row.comment)
    if s.sentiments >= 0.5:
        semantic.append('正面')
    else:
        semantic.append('負面')
 
data['semantic'] = semantic
data
```

```python
result = data.groupby('title')['semantic'].value_counts().to_frame()
result
```
